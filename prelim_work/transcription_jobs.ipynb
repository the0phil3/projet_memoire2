```python
jpg_list = []
for subdir, dirs, files in os.walk("/Users/Theo/Desktop/14-18/transcription/total_copy/"):
    for filename in files:
        if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg'):
            jpg_list.append(filename)
            
new_df = pd.DataFrame(jpg_list, columns=['file_name'])
new_df['personne'] = np.nan
new_df['fait'] = np.nan

for i in range(len(new_df)):
    if i % 9 == 0:
        new_df.loc[i:i+8, 'personne'] = name_list[i//9]
        
new_df['labat_id'] = new_df['file_name'].apply(lambda x: re.findall(r'^\d+?(?=_)', x)[0])
new_df.loc[162] = ["815346_archives_FRAD075RM_D4R1_1685_0044_D.jpg", 'Theophile_Miailhe', None, '815346']
```

```python
new_df
```

```python
ppl = new_df['personne'].unique()

# Create a dictionary to store the new dataframes
personne_dfs = {}

# Loop through unique values and create a new dataframe for each
for personne in ppl:
    personne_dfs[personne] = new_df[new_df['personne'] == personne]
```

```python
writer = pd.ExcelWriter('/Users/Theo/Desktop/14-18/transcription/groupe_excel.xlsx', engine='openpyxl')

# Loop through dictionary and save each dataframe to a different sheet
for sheetname, df in personne_dfs.items():
    df.to_excel(writer, sheet_name=sheetname, index=False)

# Save the workbook
writer.save()
```
