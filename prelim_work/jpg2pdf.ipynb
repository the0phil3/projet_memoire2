```python
import pandas as pd 
from PIL import Image
import os
```

```python
path = "/Users/Theo/Desktop/14-18/tirage/"
os.chdir(path)
```

```python
def jpeg_to_pdf(directory):
    for subdir, dirs, files in os.walk(directory):
        images = []
        for filename in files:
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                image = Image.open(os.path.join(subdir, filename))
                images.append(image)
        
        if len(images) > 0:
            output_pdf = os.path.join(subdir, "result.pdf")
            images[0].save(output_pdf, "PDF" ,resolution=100.0, save_all=True, append_images=images[1:])
            print(f"Converted images in folder {subdir} to PDF and saved it as {output_pdf}.")
```

```python
jpeg_to_pdf(path)
```

```python
def rename_pdfs(directory):
    for subdir, dirs, files in os.walk(directory):
        jpeg_file = None
        for filename in files:
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                jpeg_file = filename
                break
        
        for filename in files:
            if filename.endswith(".pdf") and jpeg_file is not None:
                pdf_file = os.path.join(subdir, filename)
                jpeg_file = os.path.splitext(jpeg_file)[0] + ".pdf"
                new_pdf_file = os.path.join(subdir, jpeg_file)
                os.rename(pdf_file, new_pdf_file)
                print(f"Renamed {pdf_file} to {new_pdf_file}.")
```

```python
rename_pdfs(path)
```
