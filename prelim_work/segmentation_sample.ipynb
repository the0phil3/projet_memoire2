```python
def delete_pdfs(directory):
    for subdir, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith(".pdf"):
                pdf_file = os.path.join(subdir, filename)
                os.remove(pdf_file)
                print(f"Deleted {pdf_file}.")
```

```python
delete_pdfs(path)
```
