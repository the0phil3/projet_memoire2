```python
def copy_jpegs_to_folder_with_foldername(directory_path, output_folder_path, folder_names):
    # Create the output folder if it doesn't already exist
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)

    # Loop over the subdirectories in the input directory
    for subdir_name in os.listdir(directory_path):
        subdir_path = os.path.join(directory_path, subdir_name)

        # If the subdirectory is a directory and its name is in the list of folder names,
        # loop over its files and copy any JPEG files to the output directory while adding the folder name to the filename
        if os.path.isdir(subdir_path) and subdir_name in folder_names:
            for filename in os.listdir(subdir_path):
                if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg'):
                    source_path = os.path.join(subdir_path, filename)
                    destination_filename = subdir_name + '_' + filename
                    destination_path = os.path.join(output_folder_path, destination_filename)
                    shutil.copyfile(source_path, destination_path)
                    print(f"Copied {source_path} to {destination_path}.")
```

```python
copy_jpegs_to_folder_with_foldername('/Users/Theo/Desktop/new_tirage/tirage/', '/Users/Theo/Desktop/theo/', theo)
```
